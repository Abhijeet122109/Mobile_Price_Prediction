Split the dataset into the independent and dependent features,
Then split the dataset into the training and testing to evaluate the model using the train_test_split method

```python
X = train.drop('price_range',axis=1)
y = train['price_range']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,y,test_size=0.1,random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
test = sc.transform(test)
```

Loading the Decision Tree Classifier from sklearn library and define the DecisionTreeClassifier and train with the X_train and Y_train dataset. Then test the model using the X_test dataset. Then check the accuracy score of the Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train , Y_train)
```

```python
pred = dtc.predict(X_test)
pred
```

Loading the Support Vector Classifier and define the SVC and train with the X_train and Y_train dataset. Then test the model using the X_test dataset. Then check the accuracy score of the Support Vector Classifier

```python
from sklearn.metrics import accuracy_score, confusion_matrix
dtc_acc = accuracy_score(pred,Y_test)
print(dtc_acc)
print(confusion_matrix(pred,Y_test))
```

```python
from sklearn.svm import SVC
knn=SVC()
knn.fit(X_train,Y_train)
```

```python
pred1 = knn.predict(X_test)
pred1
```

Loading the Logistic Regression and define the LogisticRegression and train with the X_train and Y_train dataset. Then test the model using the X_test dataset. Then check the accuracy score of the Logistic Regression

```python
from sklearn.metrics import accuracy_score
svc_acc = accuracy_score(pred1,Y_test)
print(svc_acc)
print(confusion_matrix(pred1,Y_test))
```

```python
from sklearn.linear_model import LogisticRegression  # its a classification
lr=LogisticRegression()
lr.fit(X_train,Y_train)
```

```python
pred2 = lr.predict(X_test)
pred2
```

```python
from sklearn.metrics import accuracy_score
lr_acc = accuracy_score(pred2,Y_test)
print(lr_acc)
print(confusion_matrix(pred2,Y_test))
```

Now visualize the accuracy score using the bar plot method of matplotlib. The final best performing model is Logistic Regression. Then test Logistic Regression with a separate test dataset.

```python
plt.bar(x=['dtc','svc','lr'],height=[dtc_acc,svc_acc,lr_acc])
plt.xlabel("Algorithms")
plt.ylabel("Accuracy Score")
plt.show()
```
